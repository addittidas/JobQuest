# Data Processing

```python
import pandas as pd
data_ds = pd.read_csv("data-science-skills_csv.csv")
data_ds[0:6]
```

```python
# Remove duplicate jobs
data_ds = data_ds.drop_duplicates(subset=['Job_Title', 'Company', 'City'])
```

```python
# Extract minimum and maximum values of salary
data_ds[['min_salary', 'max_salary']] = data_ds['Salary'].str.replace(',', '').str.extract(r'(\d+)-(\d+)')
data_ds['min_salary'] = data_ds['min_salary'].astype(float)
data_ds['max_salary'] = data_ds['max_salary'].astype(float)
```

```python
# Extract minimum and maximum values of experience
data_ds[['min_exp', 'max_exp']] = data_ds['Experience'].str.extract(r'(\d+)-(\d+)')
data_ds['min_exp'] = data_ds['min_exp'].astype(float)
data_ds['max_exp'] = data_ds['max_exp'].astype(float)
```

```python
data_ds.head()
```

```python
# Check for unique values in min_salary and max_salary columns
print(data_ds['min_salary'].unique())
print(data_ds['max_salary'].unique())
```

```python
# Replace with the median value of each column
data_ds['min_salary'] = data_ds['min_salary'].fillna(data_ds['min_salary'].median())
data_ds['max_salary'] = data_ds['max_salary'].fillna(data_ds['max_salary'].median())
```

```python
data_ds.head()
```

# Text Processing

```python
data_ds = data_ds.apply(lambda x: x.astype(str).str.lower())
```

```python
data_ds.City=[location.split(",") for location in data_ds.City]
```

```python
data_ds.head()
```

```python
# Save it as a .csv file
data_ds.to_csv('new-data-science-skills.csv', index=False)
```

# Data Visualization

```python
import matplotlib.pyplot as plt

data_ds.City.apply(pd.Series).stack().str.strip().value_counts()[:10].plot.pie(figsize=(7,7),startangle=50,autopct='%1.1f%%',fontsize=15)
plt.title("Location Wise Data scientist Jobs",fontsize=30)

#https://medium.com/@krishnakummar/donut-chart-with-python-matplotlib-d411033c960b
centre_circle = plt.Circle((0,0),0.72,color='gray', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.show()
```

```python
data_ds["Experience"].value_counts()[:10].plot.bar(figsize=(8,5),fontsize=13,color="b")
plt.xlabel("Experience",fontsize=18)
plt.ylabel("No.of Vacancies",fontsize=18)
plt.show()
```

```python
# Step 1: Define keywords for education levels
education_keywords = {
    'UG': ['b.tech', 'b.e.', 'b.sc.'],  # Include other undergraduate degrees as needed
    'PG': ['m.tech', 'm.s.'],  # Include other postgraduate degrees as needed
    'Doctorate': ['phd', 'doctorate', 'd.sc.'],
    # Add more categories if needed
}

# Step 2: Create a new column to categorize education levels based on keywords
def categorize_education(education):
    for category, keywords in education_keywords.items():
        if any(keyword in education for keyword in keywords):
            return category
    return 'Other'  # For any education level not specified

data_ds['Education_Category'] = data_ds['Education'].apply(categorize_education)

# Step 3: Count occurrences of each education category
education_counts = data_ds['Education_Category'].value_counts()

# Step 4: Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Education Levels", fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
data_ds['min_salary'] = pd.to_numeric(data_ds['min_salary'], errors='coerce')  
data_ds['max_salary'] = pd.to_numeric(data_ds['max_salary'], errors='coerce')  
data_ds['Average_Salary'] = (data_ds['min_salary'] + data_ds['max_salary']) / 2
```

```python
# Step 3: Plot the Average Salary
plt.figure(figsize=(10, 6))
data_ds['Average_Salary'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Average Salaries", fontsize=20)
plt.xlabel("Average Salary", fontsize=18)
plt.ylabel("Number of Vacancies", fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid(axis='y')  # Add grid lines for better readability
plt.show()
```

```python
import ast
data_ds['Skills'] = data_ds['Skills'].apply(ast.literal_eval)

individual_skills = data_ds['Skills'].explode()  # This will create a Series where each skill is an individual entry

# Step 4: Count occurrences of each skill
skill_counts = individual_skills.value_counts()

# Step 5: Plot the top 32 skills
skill_counts[:32].plot(kind="bar", figsize=(18, 6), fontsize=15, color="r")
plt.xticks(rotation=50, ha='right')
plt.title("Top Skills for Data Science", fontsize=25)  # Title for the plot
plt.ylabel("No. of Vacancies", fontsize=20)
plt.xlabel("Top Skills for Data Science", fontsize=25)
plt.show()
```

# Data Analysis of skills

```python
#lets convert the skills column into a seperate DataFrame, which makes things easy for preo processing
only_skills=pd.DataFrame(data_ds.Skills.apply(pd.Series).stack().value_counts()).reset_index()
only_skills.columns=["Skills","count"]
only_skills.head()
```

```python
#lets first find the top skills for Data science in general
datascience={}
datascience['statistics']=only_skills["count"][only_skills['Skills'].str.contains('stat', regex=True)].sum()
datascience['machine_learning']=only_skills["count"][only_skills['Skills'].str.contains('machine |^ml', regex=True)].sum()
datascience['data_analysis']=only_skills["count"][only_skills['Skills'].str.contains('data ana', regex=True)].sum()
datascience['data_mining']=only_skills["count"][only_skills['Skills'].str.contains('mining', regex=True)].sum()
datascience['nlp']=only_skills["count"][only_skills['Skills'].str.contains('nlp|natural', regex=True)].sum()
datascience['computer_vision']=only_skills["count"][only_skills['Skills'].str.contains('computer vision', regex=True)].sum()
datascience['deep_learning']=only_skills["count"][only_skills['Skills'].str.contains('deep learning', regex=True)].sum()
datascience['big_data']=only_skills["count"][only_skills['Skills'].str.contains('big', regex=True)].sum()
from operator import itemgetter
datascience=dict(sorted(datascience.items(), key=itemgetter(1),reverse=True))
datascience
```

```python
plt.bar(datascience.keys(),datascience.values(),color=["#FE6363","#66FA72","#6B89F9","c","y","pink","#63FCFE","#DD63FE"])
plt.xticks(rotation=80,fontsize=15)
plt.title("Must have skills",fontsize=20)
plt.show()
```

```python
languages={}
languages["python"]=only_skills["count"][only_skills['Skills'].str.contains('python', regex=True)].sum()
languages["r"]=only_skills["count"][only_skills['Skills'].str.contains('^r$', regex=True)].sum()
languages["matlab"]=only_skills["count"][only_skills['Skills'].str.contains('matlab', regex=True)].sum()
languages["java"]=only_skills["count"][only_skills['Skills'].str.contains('java$', regex=True)].sum()
languages["c++"]=only_skills["count"][only_skills['Skills'].str.contains('c\+', regex=True)].sum()
languages["sas"]=only_skills["count"][only_skills['Skills'].str.contains('sas', regex=True)].sum()

#to identify the Sql first and then seperate the nosql from the list
sql=only_skills[only_skills['Skills'].str.contains('sql', regex=True)]
languages["sql"]=sql["count"][~sql['Skills'].str.contains('no', regex=True)].sum()

#to sort the dictionary
languages=dict(sorted(languages.items(), key=itemgetter(1),reverse=True))
```

```python
plt.bar(languages.keys(),languages.values(),color=["r","g","b","c","y","pink","m"])
plt.xticks(rotation=45,fontsize=15)
plt.title("Programming languages for Data science",fontsize=18)
plt.show()
```

```python
frameworks={}
frameworks['tensorflow']=only_skills["count"][only_skills['Skills'].str.contains('tensor', regex=True)].sum()
frameworks['keras']=only_skills["count"][only_skills['Skills'].str.contains('keras', regex=True)].sum()
frameworks['pytorch']=only_skills["count"][only_skills['Skills'].str.contains('torch', regex=True)].sum()
plt.bar(frameworks.keys(),frameworks.values(),color=["g","b","c"],width=.5)
plt.xticks(rotation=70,fontsize=15)
plt.title("Deep learning Frameworks",fontsize=20)
plt.show()
```

```python
# tools
tools={}
tools["tableau"]=only_skills["count"][only_skills['Skills'].str.contains('tableau', regex=True)].sum()
tools["power_bi"]=only_skills["count"][only_skills['Skills'].str.contains('power bi', regex=True)].sum()


plt.bar(tools.keys(),tools.values(),color=["orange","blue"],width=(0.4))
plt.xticks(rotation=45,fontsize=15)
plt.title("Visualization Tools",fontsize=20)
plt.show()
```
