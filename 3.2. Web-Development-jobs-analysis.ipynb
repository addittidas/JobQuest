# Data Processing

```python
import pandas as pd
data_webdev = pd.read_csv("web-dev-skills_csv.csv")
data_webdev[0:6]
```

```python
# Remove duplicate jobs
data_webdev = data_webdev.drop_duplicates(subset=['Job_Title', 'Company', 'City'])
```

```python
# Extract minimum and maximum values of salary
data_webdev[['min_salary', 'max_salary']] = data_webdev['Salary'].str.replace(',', '').str.extract(r'(\d+)-(\d+)')
data_webdev['min_salary'] = data_webdev['min_salary'].astype(float)
data_webdev['max_salary'] = data_webdev['max_salary'].astype(float)
```

```python
# Extract minimum and maximum values of experience
data_webdev[['min_exp', 'max_exp']] = data_webdev['Experience'].str.extract(r'(\d+)-(\d+)')
data_webdev['min_exp'] = data_webdev['min_exp'].astype(float)
data_webdev['max_exp'] = data_webdev['max_exp'].astype(float)
```

```python
data_webdev.head()
```

```python
# Check for unique values in min_salary and max_salary columns
print(data_webdev['min_salary'].unique())
print(data_webdev['max_salary'].unique())
```

```python
# Replace with the median value of each column
data_webdev['min_salary'] = data_webdev['min_salary'].fillna(data_webdev['min_salary'].median())
data_webdev['max_salary'] = data_webdev['max_salary'].fillna(data_webdev['max_salary'].median())
```

```python
data_webdev.head()
```

# Text Processing

```python
data_webdev = data_webdev.apply(lambda x: x.astype(str).str.lower())
```

```python
data_webdev.City=[location.split(",") for location in data_webdev.City]
```

```python
data_webdev.head()
```

```python
# Save it as a .csv file
data_webdev.to_csv('new-web-dev-skills.csv', index=False)
```

# Data Visualization

```python
import matplotlib.pyplot as plt

data_webdev.City.apply(pd.Series).stack().str.strip().value_counts()[:10].plot.pie(figsize=(7,7),startangle=50,autopct='%1.1f%%',fontsize=15)
plt.title("Location Wise Web Development Jobs",fontsize=30)

#https://medium.com/@krishnakummar/donut-chart-with-python-matplotlib-d411033c960b
centre_circle = plt.Circle((0,0),0.72,color='gray', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.show()
```

```python
data_webdev["Experience"].value_counts()[:10].plot.bar(figsize=(8,5),fontsize=13,color="b")
plt.xlabel("Experience",fontsize=18)
plt.ylabel("No.of Vacancies",fontsize=18)
plt.show()
```

```python
# Step 1: Define keywords for education levels
education_keywords = {
    'UG': ['b.tech', 'b.e.', 'b.sc.'],  # Include other undergraduate degrees as needed
    'PG': ['m.tech', 'm.s.'],  # Include other postgraduate degrees as needed
    'Doctorate': ['phd', 'doctorate', 'd.sc.'],
    # Add more categories if needed
}

# Step 2: Create a new column to categorize education levels based on keywords
def categorize_education(education):
    for category, keywords in education_keywords.items():
        if any(keyword in education for keyword in keywords):
            return category
    return 'Other'  # For any education level not specified

data_webdev['Education_Category'] = data_webdev['Education'].apply(categorize_education)

# Step 3: Count occurrences of each education category
education_counts = data_webdev['Education_Category'].value_counts()

# Step 4: Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Education Levels", fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
data_webdev['min_salary'] = pd.to_numeric(data_webdev['min_salary'], errors='coerce')  
data_webdev['max_salary'] = pd.to_numeric(data_webdev['max_salary'], errors='coerce')  
data_webdev['Average_Salary'] = (data_webdev['min_salary'] + data_webdev['max_salary']) / 2
```

```python
# Step 3: Plot the Average Salary
plt.figure(figsize=(10, 6))
data_webdev['Average_Salary'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Average Salaries", fontsize=20)
plt.xlabel("Average Salary", fontsize=18)
plt.ylabel("Number of Vacancies", fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid(axis='y')  # Add grid lines for better readability
plt.show()
```

```python
import ast
data_webdev['Skills'] = data_webdev['Skills'].apply(ast.literal_eval)

individual_skills = data_webdev['Skills'].explode()  # This will create a Series where each skill is an individual entry

# Step 4: Count occurrences of each skill
skill_counts = individual_skills.value_counts()

# Step 5: Plot the top 32 skills
skill_counts[:32].plot(kind="bar", figsize=(18, 6), fontsize=15, color="r")
plt.xticks(rotation=50, ha='right')
plt.title("Top Skills for Web Development", fontsize=25)  # Title for the plot
plt.ylabel("No. of Vacancies", fontsize=20)
plt.xlabel("Top Skills for Web Development", fontsize=25)
plt.show()
```

# Data Analysis of skills

```python
#lets convert the skills column into a seperate DataFrame, which makes things easy for preo processing
only_skills=pd.DataFrame(data_webdev.Skills.apply(pd.Series).stack().value_counts()).reset_index()
only_skills.columns=["Skills","count"]
only_skills.head()
```

```python
#lets first find the top skills for Web Development in general
wd_skills={}
wd_skills['html']=only_skills["count"][only_skills['Skills'].str.contains('html', regex=True)].sum()
wd_skills['css']=only_skills["count"][only_skills['Skills'].str.contains('css', regex=True)].sum()
wd_skills['javascript']=only_skills["count"][only_skills['Skills'].str.contains('java', regex=True)].sum()
wd_skills['php']=only_skills["count"][only_skills['Skills'].str.contains('php', regex=True)].sum()
wd_skills['mysql']=only_skills["count"][only_skills['Skills'].str.contains('mysql | sql', regex=True)].sum()
wd_skills['ajax']=only_skills["count"][only_skills['Skills'].str.contains('ajax', regex=True)].sum()
wd_skills['python']=only_skills["count"][only_skills['Skills'].str.contains('python', regex=True)].sum()
wd_skills['bootstrap']=only_skills["count"][only_skills['Skills'].str.contains('bootstrap', regex=True)].sum()
from operator import itemgetter
wd_skills=dict(sorted(wd_skills.items(), key=itemgetter(1),reverse=True))
wd_skills
```

```python
plt.bar(wd_skills.keys(),wd_skills.values(),color=["#FE6363","#66FA72","#6B89F9","c","y","pink","#63FCFE","#DD63FE"])
plt.xticks(rotation=80,fontsize=15)
plt.title("Must have skills",fontsize=20)
plt.show()
```

```python
# Languages specific to web development
languages = {
    "html": 0,
    "css": 0,
    "javascript": 0,
    "typescript": 0,
    "php": 0,
    "ruby": 0,
    "python": 0,
    "java": 0,
    "c#": 0,
    "swift": 0,
    "go": 0,
    "kotlin": 0,
    "scala": 0,
    "elixir": 0,
    "rust": 0,
    "sql": 0,
    "graphql": 0,
}

# Count languages
for lang in languages.keys():
    languages[lang] = individual_skills.str.contains(lang, regex=True, case=False).sum()

# Sort the languages dictionary
languages = dict(sorted(languages.items(), key=itemgetter(1), reverse=True))
```

```python
# Plotting the programming languages
plt.bar(languages.keys(), languages.values(), color=["r", "g", "b", "c", "purple", "orange", "pink", "cyan", "brown", "yellow", "grey", "lightblue", "lightgreen", "darkorange", "gold", "lightcoral", "violet"])
plt.xticks(rotation=45, fontsize=15)
plt.title("Programming Languages for Web Development", fontsize=18)
plt.show()
```

```python
# Frameworks specific to web development
frameworks = {
    'angularjs': 0,
    'react': 0,
    'vue': 0,
    'django': 0,
    'flask': 0,
    'spring': 0,
    'laravel': 0,
}

# Count frameworks
for framework in frameworks.keys():
    frameworks[framework] = individual_skills.str.contains(framework, regex=True, case=False).sum()

# Sort frameworks dictionary
frameworks = dict(sorted(frameworks.items(), key=itemgetter(1), reverse=True))
```

```python
# Plotting the frameworks
plt.bar(frameworks.keys(), frameworks.values(), color=["blue", "green", "purple", "orange", "red", "brown", "pink"], width=0.5)
plt.xticks(rotation=70, fontsize=15)
plt.title("Popular Web Development Frameworks", fontsize=20)
plt.show()
```
